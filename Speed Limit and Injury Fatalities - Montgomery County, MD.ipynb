#### Step 3: Finding the PERCENTAGE of the predictor for possible injuries  per each speed limit zone.

Use the "Apply-Combine" strategy to combine this question below

##### What is the rate of total percentage that people are likely to encounter injuries on EACH of their speed limit zones as reported from the Top 5?

HINT: Use the statement that I wrote that on the previous step that gives an percentage overall summary of these reports.

Convert this to a csv file when finished - should return the table

```python
import pandas as pd

#make a def functin
def percentage_amount(reported):
    #make a list
    reported_fatalities_speed_limit = []

    for speedlimit, speedlimitdata in car_accidents.groupby('Speed Limit'):
        #2). Split the column above by calculating the mean of the total amount of injuries per each speed limit zone.
        ## Give me a representation of the speed limit zone 
        ## (HINT use the instance code of the speedlimitdata - gives a range of values from the cols in the dataframe)
        speed_limit = int(speedlimit)
        ## Calculate the total reported fatalities per each speed limit zone
        mean_reported = sum(speedlimitdata["Injury Severity"].value_counts())
        ## Total Reported as defined by the this dataframe
        total_reported = sum(car_accidents["Injury Severity"].value_counts())
        ## Divide by the reported to find the percentage
        percentage = (mean_reported)/(total_reported)

        #3). Summarize it
        summarize = {
            'Speed Limit Zone': speed_limit,  #Speed Limit
            'Percentage of Fatalities Reported ': percentage  #Percentage of Fatalities Reported
        }
        #4). Convert it to a new dataframe from what you summarize in 3
        reported_fatalities_speed_limit.append(summarize)

    #5). Convert it to a dataframe
    summarize = pd.DataFrame(reported_fatalities_speed_limit)
    #6). Save it to a csv file
    summarize.to_csv('/Users/neilvu/Documents/Data Analyst/Projects/INST126/Summary of Total Reported Fatalities Per Each Speed Limit Zone.csv')
    #7). Return It
    return summarize

#test it here
result = percentage_amount(car_accidents)
result

## The data should note this, for each of the speed limit zone encounter, there would be a {percentage} percantage chance that he/she likely to be 
## injured in this {speed limit type} zone.
```
